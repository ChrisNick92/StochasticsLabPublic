# Το ενιαίο πρόγραμμα - Η βασική κλάση

Παρακάτω δίνεται ο πηγαίος κώδικας της βασικής κλάσης που θα χρησιμοποιηθεί για την απαντήση των παραπάνω ερωτημάτων. Η λειτουργία της κάθε συνάρτησης δίνεται μαζί με την απάντηση του αντίστοιχου ερωτήματος. 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Only for better visualizations
import pandas as pd
import imageio
import os

sns.set_theme()
```

```python
class linear_regressor():
    def __init__(self, data):
        self.data = data      # data is a numpy array with shape (num_samples, 2)
        self.x = data[:,0]    # first column is the independent variable
        self.y = data[:,1]    # second column is the dependent variable
        self.w = np.array([0,0]) # weights of the regression line
        self.x_norm = None
        self.y_norm = None
        
    def correlation(self):
        return np.corrcoef(self.x, self.y)[0,1]
    
    def plot(self, title = "x vs y", x = "x", y = "y", with_line = False,
             save = False, pngname = None, norm = False, path = None):
        
        if norm:
            self.x_plot, self.y_plot = self.x_norm, self.y_norm
        else:
            self.x_plot, self.y_plot = self.x, self.y
        
        fig, ax = plt.subplots(figsize = (12,8))
        ax.scatter(x = self.x_plot, y = self.y_plot)
        ax.set_xlabel(x)
        ax.set_ylabel(y)
        ax.set_title(title)
        
        if with_line:
            ax.plot(self.x_plot, np.dot(np.stack([np.array(self.x_plot.shape[0]*[1]), self.x_plot], axis = 1), self.w), 
            label = "regression line: " + f"y = {self.w[0]:.3f} + ({self.w[1]:.3f}*x)", color = "green")
            ax.legend()
        
        if save:
            plt.savefig(os.path.join(path,pngname + ".png"))
            plt.close()
        else:
            sns.set_theme()
            plt.show()
            
    def normalization(self):
        self.x_norm = (self.x - self.x.min())/(self.x.max()- self.x.min())
        self.y_norm = (self.y - self.y.min())/(self.y.max()- self.y.min())
    
    def data_splitter(self, train_ratio = 0.7, seed = 42, norm = True):
        np.random.seed(seed = seed)
        p = np.random.permutation(self.y.shape[0]) # Generate a random permutation according to seed
        point = int(train_ratio*self.y.shape[0])
        
        if norm == False:
            features, labels = self.x[p], self.y[p]
        else:
            features, labels = self.x_norm[p], self.y_norm[p]

        self.X_train, self.y_train = features[:point], labels[:point]
        self.X_test, self.y_test = features[point:], labels[point:]
        
        return self.X_train, self.y_train, self.X_test, self.y_test
    
    def fit(self, X_train, y_train, lr = 0.1, epochs = 20, seed = 42, gif = False,
            pngname = None, path = None, title = None, x_label = "x", y_label = "y"):
        np.random.seed(seed = seed)
        self.epochs = epochs
        self.lr = lr
        
        # Training starts here
        for epoch in range(self.epochs):
            p = np.random.permutation(X_train.shape[0])
            X_train, y_train = X_train[p], y_train[p]
            
            # Weight Update
            for x,y in np.array(list(zip(X_train, y_train))):
                w_new = self.w + self.lr*(y - np.dot(self.w,np.array([1,x])))*np.array([1,x])
                self.w = w_new
                
            # The following commands generate the gif file
            if gif:
                self.plot(x = x_label, y = y_label, with_line = True, save = True, norm = True,
                          path = path, title = title + str(epoch+1),
                          pngname = pngname +"-" + str((epoch+1)//10) + str(epoch+1))
    
    def predict(self, X_test):
        return self.w[0] + self.w[1]*X_test
    
    def MSE(self, X_test, y_test):
        preds = self.predict(X_test)
        return (1/len(y_test))*np.sum((y_test - preds)**2)
    
```

Και οι δύο παραπάνω τιμές είναι ιδιαίτερα ικανοποιητικές για να θεωρήσουμε ότι υπάρχει έντονη γραμμική εξάρτηση (συσχέτιση) μεταξύ της ανεξάρτητης και της εξαρτημένης μεταβλητής.

Ο συντελεστής συσχέτισης κατά Pearson $r_{xy}$ για δύο διανύσματα $x = (x_1,\dots,x_n),\, y = (y_1,\dots,y_n)$ δίνεται από τη σχέση 
$$\tag{2.1} r_{xy} = \frac{\sum_{i=1}^n(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^n(x_i- \bar{x})^2}\cdot \sqrt{\sum_{i=1}^n (y_i - \bar{y})^2}},$$
όπου $\bar{x},\bar{y}$ ο δειγματικός μέσος των δύο διανυσμάτων $x,y$. Ο συντελεστής συσχέτισης ικανοποιεί την ανίσωση $-1\leq r_{x,y}\leq 1$. Τιμές που βρίσκονται κοντά στα "άκρα" ($\pm 1$) υποδεικνύουν έντονη γραμμική εξάρτηση μεταξύ των τιμών του διανύσματος $x$ και των αντίστοιχων τιμών του διανύσματος $y$. Αν η τιμή βρίσκεται κοντά στο $1$, τότε έχουμε υψηλή γραμμική θετική γραμμική εξάρτηση, δηλαδή, όσο αυξάνουν οι τιμές του $x$ αναμένουμε να δούμε και αντίστοιχη αύξηση στις τιμές του διανύσματος $y$. Αντιθέτως, τιμές κοντά στο $-1$ υποδηλώνουν έντονη αρνητική γραμμική εξάρτηση, το οποίο σημαίνει ότι όσο αυξάνονται οι τιμές του διανύσματος $x$ αναμένουμε να δούμε ανάλογη μείωση των τιμών του διανύσματος $y$.

> *Κανονικοποιήστε τα δεδομένα που παρέχονται με τη μέθοδο min-max normalization. Περισσότερες πληροφορίες για τη μέθοδο αυτή μπορούν να βρεθούν [εδώ](#https://en.wikipedia.org/wiki/Feature_scaling).*

Σε αυτό το σημείο μέσω της συνάρτησης *normalization*, η οποία υλοποιείται μέσω της βασική μας κλάσης, κανονικοποιούμε τα δεδομένα μέσω της μεθόδου *min-max normalization*. Η μέθοδος αυτή συνοψίζεται ως εξής: Αν $x = (x_1,\dots,x_n)$ είναι ένα διάνυσμα πραγματικών αριθμών, τότε η μέθοδος επιστρέφει ένα διάνυσμα $x' = (x_1',\dots,x_n')$, όπου 
$$\tag{4.1} x_i' = \frac{\max\limits_{1\leq i \leq n}{x_i} - x_i}{\max\limits_{1\leq i \leq n} x_i - \min\limits_{1\leq i \leq n}x_i}.$$

Παρακάτω βλέπουμε τα κανονικοποιήμενα δεδομένα που προκύπτουν από τα δεδομένα του 1ου dataset που αφορά τα βίντεο στο YouTube.

Ο χωρισμός των δεδομένων σε *train* και *test* set γίνεται μέσω της συνάρτησης *data_splitter* της βασικής κλάσης. Η συνάρτηση δέχεται ως παραμέτρο το ποσοστό του συνολικού dataset που θέλουμε να έχει ως μέγεθος το σύνολο εκπαίδευσης. > *Υπολογίστε τις παραμέτρους του μοντέλου linear regression, χρησιμοποιώντας τη μέθοδο Stochastic Gradient Descent (SGD). Για το σκοπό αυτό, ο χρήστης θα πρέπει να ορίζει το ρυθμό μάθησης (learning rate) και το συνολικό αριθμό των εποχών (epochs) εκπαίδευσης. Να θεωρήσετε ότι οι αρχικές τιμές των παραμέτρων είναι μηδενικές.*

Η εκτίμηση των παραμέτρων του μοντέλου *linear_regression* με τη μέθοδο *Stochastic Gradient Descent (SGD)* υλοποιείται μέσω της συνάρτησης *fit* της βασικής κλάσης. Η συνάρτηση δέχεται ως παραμέτρους τα χαρακτηριστικά του συνόλου εκπαίδευσης (`X_train`), τις ταμπέλες που αντιστοιχούν στο σύνολο εκπαίδευσης (`y_train`), τον ρυθμό εκμάθησης (`lr`), το σύνολο των εποχών (`epochs`) και το φίτρο (`seed`) που αντιστοιχεί στην τυχαία αρχικοποιήση για το ανακάτεμα του συνόλου εκπαίδευσης πριν από κάθε εποχή.

Η μέθοδος λειτουργεί ως εξής: Αρχικά πριν την είσοδο σε μια νεα εποχή το σύνολο εκπαίδευσης ανακατεύεται με τυχαίο τρόπο. Ύστερα, περνώντας ένα - ένα τα δεδομένα η ανανέωση των βαρών απ' το δεδομένο $i$ στο $i+1$ γίνεται με βάση την εξίσωση
$$\tag{6.1} w_{i+1} = w_i + lr\cdot \bigl(y_i - w_i^T\cdot [1,x_i]\bigr)\cdot [1,x_i],$$
όπου $w_i$ είναι το διάνυσμα βαρών στην $i-$οστή επανάληψη, $x_i$ είναι η τιμή που αντιστοιχεί στην ανεξάρτητη μεταβλητή της $i$-οστής παρατήρησης (μετά το ανακάτεμα), $y_i$ είναι η τιμή της εξαρτημένης μεταβλητής και $lr$ είναι ο ρυθμός εκμάθησης. Παρακάτω βλέπουμε το διάνυσμα βαρών μετά την εκτίμηση των παραμέτρων για ρυθμό εκμάθησης $0.1$ και $20$ εποχές. Επίσης, βλέπουμε και ένα σχήμα με τα δεδομένα και την αντίστοιχη προσαρμοσμένη ευθεία μετά την εκτίμηση των παραμέτρων.

> *Θα υπολογίστε το μέσο τετραγωνικό σφάλμα (mean squared error) των προβλέψεων πάνω στο test set. Περισσότερες πληροφορίες μπορούν να βρεθούν [εδώ](#https://en.wikipedia.org/wiki/Mean_squared_error).*

Σε αυτό το σημείο έχοντας υπολογίσει τις παραμέτρους των δύο ευθειών για τα δύο σύνολα δεδομένων υπολογίζουμε το μέσο τετραγωνικό σφάλμα το οποίο μας δίνει ένα κριτήριο αξιολόγησης των δύο μοντέλων. Αν συμβολίζουμε με $y = (y_1,\dots,y_n)$ το διάνυσμα των πραγματικών τιμών (*true labels*) και με $\hat{y} = (\hat{y_1},\dots,\hat{y_n})$ το διάνυσμα των προβλέψεων (*predicted labels*), τότε το μέσο τετραγωνικό σφάλμα δίνεται από τη σχέση
$$\tag{7.1} MSE(y,\hat{y}) = \frac{1}{n}\sum_{i=1}^n(y_i - \hat{y_i})^2.$$

```python
print(f"- The mean square error for the YouTube dataset is: {lmA.MSE(X_testA, y_testA):.8f}")
print(f"- The mean square error for the second dataset is: {lmB.MSE(X_testB, y_testB):.8f}")
```

```python
import imageio
import os

def generate_line_gif(data, title, pngname, fps = 1, 
                      filename = "TempFile", epochs = 20, lr = 0.1, split_ratio = 0.7,
                     x = "x", y = "y"):
    
    regressor = linear_regressor(data)
    
    try:
        # Create Directory
        path = os.getcwd()
        os.mkdir(filename)
        print("Directory" , filename ,  "Created ")
        
        regressor.normalization()
        X_train, y_train, X_test, y_test = regressor.data_splitter(train_ratio = split_ratio)
        regressor.fit(X_train, y_train, lr = lr, epochs = epochs,
                     gif = True, x_label = x, y_label = y, pngname = pngname, path = os.path.join(os.path.join(path, filename)),
                     title = title)
        
        png_dir = os.path.join(path, filename)
        images = []
        for file_name in sorted(os.listdir(png_dir)):
            if file_name.endswith('.png'):
                file_path = os.path.join(png_dir, file_name)
                images.append(imageio.imread(file_path))
        imageio.mimsave(pngname +".gif", images, fps = fps)
        
      # Remove the specified  
      # file path
        names = os.listdir(png_dir)
        for fname in sorted(names, key=lambda x:int(x.split('-')[-1].split('.')[0])):
            png_path = os.path.join(path, png_dir)
            if fname.endswith(".png"):
                os.remove(os.path.join(png_path, fname))
        try: 
            os.rmdir(filename) 
            print("% s removed successfully" % os.path.join(path, filename))
        except OSError as error: 
            print(error) 
            print("File path can not be removed") 
            
    except FileExistsError:
        print("Directory " , filename ,  " already exists")
```

><b>1.</b> *Να ταξινομήσετε τις καταστάσεις της αλυσίδας σε κλάσεις επικοινωνίας. Ποιες κλάσεις είναι ανοιχτές και ποιες κλειστές; Ποιες καταστάσεις είναι παροδικές και ποιες επαναληπτικές; Για να βοηθηθείτε μπορείτε να χρησιμοποιήσετε τον <a href="https://setosa.io/markov/index.html#%7B%22tm%22%3A%5B%5B0.5%2C0.5%2C0%2C0%2C0%2C0%2C0%2C0%5D%2C%5B0.25%2C0.75%2C0%2C0%2C0%2C0%2C0%2C0%5D%2C%5B0.25%2C0.25%2C0%2C0.125%2C0.375%2C0%2C0%2C0%5D%2C%5B0%2C0%2C0.25%2C0%2C0.75%2C0%2C0%2C0%5D%2C%5B0%2C0%2C0.2%2C0.2%2C0.2%2C0.2%2C0.2%2C0%5D%2C%5B0%2C0%2C0%2C0%2C0%2C0%2C0.5%2C0.5%5D%2C%5B0%2C0%2C0%2C0%2C0%2C0.5%2C0%2C0.5%5D%2C%5B0%2C0%2C0%2C0%2C0%2C0.5%2C0.5%2C0%5D%5D%7D">προσομοιωτή</a>, ο οποίος είναι κατάλληλα διαμορφωμένος για την άσκηση.*

<b>1.A *Κλάσεις επικοινωνίας*</b> 

- Οι <b>*κλάσεις επικοινωνίας*</b> περιέχουν τις καταστάσεις για τις οποίες υπάρχει θετική πιθανότητα να πάμε απ'τη μία κατάσταση στην άλλη και αντίστροφα σε πεπερασμένα βήματα. Παρατηρώντας τον προσομοιωτή βλέπουμε ότι οι κλάσεις επικοινωνίας που σχηματίζονται με βάση τον πίνακα $P$ είναι οι $C_1 = \{1,2\},\, C_2 = \{3,4,5\}$ και η $C_3 = \{6,7,8\}.$


- Οι <b>*ανοικτές κλάσεις*</b> είναι αυτές οι οποίες επικοινωνούν με κάποια άλλη κλάση, δηλαδή, υπάρχει θετική πιθανότητα η μαρκοβιανή αλυσίδα να φύγει από την κλάση και να πάει σε κάποια άλλη. Απ' το σχήμα του προσομοιώτη βλέπουμε ότι υπάρχει μια ανοικτή κλάση, η $C_2 = \{3,4,5\}$.


- <b>*κλειστές κλάσεις*</b> είναι αυτές για τις οποίες αν η αλυσίδα βρεθεί μέσα σε αυτές τότε δεν μπορεί να φύγει από αυτήν και να πάει σε κάποια άλλη κλάση. Στο συγκεκριμένο παράδειγμα οι κλειστές κλάσεις είναι η $C_1=\{1,2\}$ και η $C_3 =\{6,7,8\}$.


- <b>*Παροδική*</b> είναι μια κατάσταση στην οποία μετά από πεπερασμένα βήματα μεταβάσεων η διαδικασία δεν επιστρέφει πότε ξανά σε αυτή. Στο παράδειγμά μας παροδικές είναι όλες οι καταστάσεις της κλάσης $C_2$, δηλαδή οι $3,4,5$.


- <b>*Επαναληπτική*</b> είναι μια κατάσταση στην οποία η αλυσίδα επιστρέφει άπειρες φορές στο διηνεκές. Στο παράδειγμά μας οι επαναληπτικές καταστάσεις είναι οι $C_1\cup C_3 = \{1,2,6,7,8\}$.

><b>2. </b>*Χρησιμοποιώντας το αποτέλεσμα του προηγούμενου ερωτήματος, να προσδιορίσετε την αναλλοίωτη κατανομή της μαρκοβιανής αλυσίδας για την περίπτωση που η αλυσίδα ξεκινάει από την κατάσταση (α) 1 και (β) 8, χωρίς τη βοήθεια υπολογιστή.*

<b>*(α) Εκκίνηση από την κατάσταση 1.*</b>

Στη περίπτωση που η αλυσίδα ξεκινάει από την κατάσταση 1 τότε αφού η 1 περιέχεται στην κλειστή κλάση $C_1$, οι μόνες καταστάσεις που θα επισκέπτεται η αλυσίδα θα είναι οι 1,2. Έτσι, σε αυτή την περίπτωση θα έχουμε μια νεα μαρκοβιανή αλυσίδα της οποίας ο πίνακας μεταβάσεων $P_1$ θα είναι ο υποπίνακας
$$P_1 = \begin{pmatrix}
1/2 & 1/2\\
1/4 & 3/4\\
\end{pmatrix},
$$
του $P$. Τώρα, αφού η κλάση $C_1$ είναι κλειστή και οι καταστάσεις της $C_1$ επικοινωνούν μεταξύ τους, η νέα μαρκοβιανή αλυσίδα θα είναι <b>μη υποβιβάσιμη</b>. Έτσι, η αναλλοίωτη κατανομή αυτής της αλυσίδας θα δίνεται μέσω της λύσης του συστήματος

\begin{align}
\begin{cases}
\frac{\pi_1}{2}+\frac{\pi_2}{4} = \pi_1\\
\frac{\pi_1}{2}+\frac{3}{4}\pi_2 = \pi_2\\
\pi_1 + \pi_2 = 1.
\end{cases}
\end{align}

Τώρα, έχουμε 

\begin{align}
\frac{\pi_1}{2}+\frac{\pi_2}{4} = \pi_1 &\Longleftrightarrow 2\pi_1 +\pi_2 = 4\pi_1\\
& \Longleftrightarrow \pi_2 = 2\pi_1.
\end{align}

Άρα, χρησιμοποιώντας και την $\pi_1+\pi_2 = 1$ θα έχουμε ότι $3\pi_1 = 1\Longrightarrow \pi_1 = 1/3$. Έτσι, παίρνουμε ότι $\pi_2 = 2/3$. Άρα εν τέλει η αναλλοίωτη κατανομή σε αυτή την περίπτωση θα είναι η $\pi = (1/3,\, 2/3,\, 0,\, 0,\, 0,\, 0,\, 0,\, 0)$.

<b>*(β) Εκκίνηση από την κατάσταση 8.*</b>

Κατά παρόμοιο τρόπο στην περίπτωση που η αλυσίδα ξεκινάει απ'την κατάσταση 8 και επειδή αυτή η κατάσταση περιέχεται στην κλειστή κλάση $C_3$ θα δημιουργείται μια ύπο-αλυσίδα της αρχικής αλυσίδας με πίνακα μεταβάσεων τον υποπίνακα
$$P_2 = \begin{pmatrix}
0 & 1/2 & 1/2\\
1/2 & 0 & 1/2\\
1/2 & 1/2 & 0\\
\end{pmatrix},
$$
του πίνακα $P$. Εφόσον η ύπο-αλυσίδα είναι μια <b>μη υποβιβάσιμη</b> αλυσίδα τότε θα υπάρχει αναλλοίωτη κατανομή η οποία δίνεται μέσω της λύσης του συστήματος

\begin{align}
\begin{cases}
\frac{\pi_7}{2}+\frac{\pi_8}{2} = \pi_6\\
\frac{\pi_6}{2}+\frac{\pi_8}{2} = \pi_7\\
\frac{\pi_6}{2}+\frac{\pi_7}{2} = \pi_8\\
\pi_6 + \pi_7 + \pi_8 = 1.
\end{cases}
\end{align}

Τώρα, γράφοντας

\begin{align}
\begin{cases}
\frac{\pi_7}{2}+\frac{\pi_8}{2} = \pi_6\\ 
\frac{\pi_6}{2}+\frac{\pi_8}{2} = \pi_7
\end{cases}
&\Longleftrightarrow
\begin{cases}
\pi_7 + \pi_8 = 2 \pi_6\\
\pi_6 + \pi_8 = 2\pi_7
\end{cases}
\Longleftrightarrow 
\begin{cases}\tag{*}
\pi_7 = 2\pi_6 - \pi_8\\
\pi_6 + \pi_8 = 2\pi_7.
\end{cases}
\end{align}

Απ'όπου καταλήγουμε στο ότι $\pi_6 + \pi_8 = 4\pi_6 - 2\pi_8 \Longleftrightarrow \pi_6 = \pi_8.$ Επομένως, από την $(*)$ καταλήγουμε στο ότι $\pi_6 = \pi_7 = \pi_8$, και επειδή $\pi_6 + \pi_7 + \pi_8 =1$ θα έχουμε ότι $\pi_6 = \pi_7 = \pi_8 =1/3$. Συνεπώς, η αναλλοίωτη κατανομή θα είναι η $\pi = (0,\,0,\,0,\,0,\,0,\, 1/3,\, 1/3,\, 1/3)$.

> <b>3.</b> *Να θεωρήσετε την περίπτωση στην οποία η αλυσίδα ξεκινάει από την κατάσταση 1. Να υποθέσετε ότι σε κάποιο βήμα, η αλυσίδα βρίσκεται στην κατάσταση 2 και ότι έχει παρέλθει το μεταβατικό φαινόμενο. Ποιος είναι ο αναμενόμενος αριθμός βημάτων που απαιτούνται για την πρώτη επιστροφή της αλυσίδας
στην κατάσταση 2; Να απαντήσετε στο ερώτημα χωρίς τη βοήθεια υπολογιστή.*

##### 1ος Τρόπος

Όπως ήδη έχουμε αναφέρει και παραπάνω, στη περίπτωση που η αλυσίδα ξεκινάει απ'τη κατάσταση 1 θα είμαστε στη περίπτωση που δημιουργείται μια ύπο-αλυσίδα με πίνακα μεταβάσεων που δίνεται από τον $P_1$. Έτσι, αφού η κλάση $C_1$ είναι κλειστή, οι μόνες πιθανές καταστάσεις της ύπο-αλυσίδας θα είναι οι $1,2$. Τώρα, αν υποθέσουμε ότι η αλυσίδα βρίσκεται στην κατάσταση $2$ για να βρούμε τον αναμενόμενο αριθμό βημάτων μέχρις ότου η αλυσίδα να επιστρέψει στην κατάσταση $2$ σκεφτόμαστε ως εξής: Η αλυσίδα θα επιστρέψει είτε σε $1$ βήμα, είτε $2$ βήματα, είτε σε $3$ βήματα κτλ. Συνεπώς, αν συμβολίσουμε με $X$ την τυχαία μεταβλητή που εκφράζει τον αριθμό βημάτων μέχρι την πρώτη επιστροφή στην κατάσταση $2$ τότε για τη μέση της τιμή θα έχουμε
\begin{align}
E[X] &= 1\cdot P\bigl(X=1\bigr)+2\cdot P\bigl(X=2\bigr)+\dots\\ &= \sum_{n=1}^{\infty}n\cdot P\bigl(X=n\bigr).
\end{align}

Τώρα, εφόσον ξεκινάμε απ'τη κατάσταση $2$, η πιθανότητα να επιστρέψουμε σε αυτή σε ένα βήμα θα είναι ίση με $P\bigl(X=1\bigr) = 3/4$. Η πιθανότητα να επιστρέψουμε σε δύο βήματα θα είναι ίση με την πιθανότητα σε ένα βήμα να πάμε στη κατάσταση $1$ επί την πιθανότητα από την κατάσταση $1$ να πάμε στη κατάσταση $2$. Δηλαδή, $P\bigl(X=2\bigr) = \frac{1}{4}\cdot \frac{1}{2}$. Κατ' αντιστοιχία, πιθανότητα η αλυσίδα να επιστρέψει μετά από $3$ βήματα στην κατάσταση $2$ θα είναι ίση με $P\bigl(X=3\bigr)=\frac{1}{4}\cdot \bigl(\frac{1}{2}\bigr)^2$ και γενικότερα σε $n$ βήματα (με $n>1$) θα είναι ίση με $P\bigl(X=n\bigr) = \frac{1}{4}\cdot \bigl(\frac{1}{2}\bigr)^{n-1}$. Άρα ο αναμενόμενος αριθμός βημάτων μέχρις ότου η αλυσίδα να επιστρέψει στην κατάσταση $2$ θα είναι ίσος με 
\begin{align}
E[X] &= \sum_{n=1}^{\infty}n\cdot P\bigl(X=n\bigr)\\
&= \frac{3}{4}+\frac{1}{4}\sum_{n=2}^{\infty}n\biggl(\frac{1}{2}\biggr)^{n-1}. \tag{1}\\
\end{align}

Τώρα, για να υπολογίσουμε το παραπάνω άθροισμα γνωρίζουμε ότι για $0<|x|<1$ ισχύει ότι
$$\sum_{n=0}^{\infty}x^n = \frac{1}{1-x}.$$
Ειδικότερα, 
$$\sum_{n=1}^{\infty}x^n = \frac{1}{1-x}-1=\frac{x}{1-x}.$$
Παραγωγίζοντας την παραπάνω δυναμοσειρά θα έχουμε ότι
\begin{align}
\sum_{n=1}^{\infty}nx^{n-1} &= \frac{1}{(1-x)^2}\implies\\
\sum_{n=2}^{\infty}nx^{n-1} &= \frac{1}{(1-x)^2} - 1 = \frac{x(2-x)}{(1-x)^2}.
\end{align}
Απ'όπου παίρνουμε για $x=1/2$ στον παραπάνω τύπο ότι
$$\sum_{n=2}^{\infty}n\biggl(\frac{1}{2}\biggr)^{n-1} = \frac{1}{2}\cdot \frac{\frac{3}{2}}{\frac{1}{4}}=3.\tag{2}$$

Επομένως, συνδυάζοντας τις $(1),(2)$ θα έχουμε ότι 
$$E[X] = \frac{3}{4} + \frac{3}{4} = \frac{6}{4} = 1.5.$$

Άρα ο αναμενόμενος αριθμός βημάτων μέχρι την πρώτη επιστροφή στην κατάσταση $2$ είναι ίσος με $1.5$ βήματα.

##### 2ος Τρόπος

Όπως έχουμε βρει στο πρώτο ερώτημα στην περίπτωση που η αλυσίδα ξεκινάει απ' την κατάσταση $1$ τότε η αλυσίδα θα έχει αναλλοίωτη κατανομή την $\pi = (1/3,\, 2/3,\, 0,\, 0,\, 0,\, 0,\, 0,\, 0)$. Έτσι, ο αριθμός αναμενόμενων βημάτων μέχρι την πρώτη επιστροφή στην κατάσταση $2$ θα είναι ίσος με $1/\pi_2 = 3/2$, το οποίο συμπίπτει με την απόδειξη του 1ου τρόπου.
